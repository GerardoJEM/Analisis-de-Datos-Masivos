```python
import pandas as pd
import numpy as np
```

### Paso 1. Carga, Limpieza y Transformacion de datos

```python
df_canada = pd.read_excel('Canada.xlsx', 
                        sheet_name = 'Canada by Citizenship',
                       skiprows = range(20),
                       skipfooter=2)
df_canada
```

```python
print(df_canada.shape)
```

```python
df_canada.drop(['AREA', 'REG' , 'DEV', 'Type', 'Coverage'], 
                axis=1, inplace = True )
df_canada.head()
```

```python
df_canada.head()
```

```python
df_canada.rename(columns = {'OdName':'Country',
                           'AreaName':'Continent',
                           'RegName':'Region'}, inplace=True)
df_canada.head()
```

```python
# verificar que todos los tipos de columna sean string
all(isinstance(column, str) for column in df_canada.columns)
```

```python
# cambiar todas las columnas a tipo String
df_canada.columns = list(map(str, df_canada.columns))
```

```python
# verificar que todos los tipos de columna sean string
all(isinstance(column, str) for column in df_canada.columns)
```

```python
df_canada.set_index('Country', inplace=True)
df_canada.head()
```

```python
# agregamos la columna total sumando la inmigracion 
#de todos los años por pais
df_canada['Total'] = df_canada.sum(axis=1)
df_canada.head()
```

```python
years = list(map(str, range(1980, 2014)))
years
```

### Parte 2. Visualizacion de datos con matplotlib

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

#### Graficos de area

```python
df_canada.sort_values(['Total'], ascending=False, 
                      axis=0, inplace=True) 
df_canada.head(5)
```

```python
df_top5 = df_canada.head()
df_top5
```

```python
# Transponer renglones y columnas
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.index = df_top5.index.map(int)
df_top5
```

```python
df_top5.plot( kind='area',
            stacked=False,
            figsize=(20,10) )
plt.title('Tendencias de inmigracion de los primeros 5 paises a Canada')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
# tranparencia
df_top5.plot( kind='area',
             alpha=0.25,
            stacked=False,
            figsize=(20,10) )
plt.title('Tendencias de inmigracion de los primeros 5 paises a Canada')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
ax = df_top5.plot(kind = 'area',
                  alpha=0.35,
                  figsize=(20,10))
ax.set_title('Tendencias de inmigracion de los primeros 5 paises a Canada')
ax.set_ylabel('Numero de inmigrantes')
ax.set_xlabel('Años')
```

### Grafico de barras

```python
df_chile = df_canada.loc['Chile', years]
df_chile.head()
```

```python
# en los diagrmas de barras el eje x se usa para eiquetar y la
# longitud de la barra corresponde al valor de la variable

df_chile.plot(kind='barh', figsize=(10,10))
plt.xlabel('Numero de inmigrantes')
plt.ylabel('Años')
plt.title('Inmigrantes chilenos a canada desde 1980 al 2013')
for index, value in enumerate(df_chile):
    label = format(int(value), ',')
    plt.annotate(label, xy=(value, index), color='black')
plt.show()
```

```python
df_mexico = df_canada.loc['Mexico', years]
df_mexico.head()
```

```python
df_mexico.plot(kind='bar', figsize=(10,6))
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.title('Inmigrantes mexicanos a canada desde 1980 al 2013')
plt.show()
```

### Graficos Circulares

```python
df_continentes = df_canada.groupby('Continent', axis=0).sum()
df_continentes
```

```python
df_continentes['Total'].plot(kind='pie',
                            figsize= (5,6),
                            autopct='%1.1f%%',
                            startangle=0,
                            shadow=True)
plt.title('Inmigracion a Canada por continente [1980-2013]')
plt.show()
```

```python
explode_list=[0.1, 0,0,0, 0.1,0.1]

df_continentes['Total'].plot(kind='pie',
                            figsize= (15,6),
                            autopct='%1.1f%%',
                            startangle=0,
                            shadow=False,
                            labels=None,
                            pctdistance=1.15,
                            explode =explode_list)
plt.title('Inmigracion a Canada por continente [1980-2013]')
plt.axis('equal')
plt.legend(labels=df_continentes.index, loc='lower left')
plt.show()
```
